# Missing Data - Initial Intuition


â€¢ Here, we don't have any missing data.

General Thumb Rules:

For features with less missing values- can use regression to predict the missing values or
fill with the mean of the values present, depending on the feature.

.For features with very high number of missing values- it is better to drop those columns
as they give very less insight on analysis.

â€¢ As there's no thumb rule on what criteria do we delete the columns with high number of
missing values, but generally you can delete the columns, if you have more than 30-40%
of missing values. But again there's a catch here, for example, Is_Car & Car_Type, People
having no cars, will obviously have Car_Type as NaN (null), but that doesn't make this
column useless, so decisions has to be taken wisely

# DATA cleaning

copy the data base

```python
telco_data = telco_base_data.copy()
```

2. Total Charges should be numeric amount. Let's convert it to numerical data type

```python
telco_data.TotalCharges = pd.to_numeric(telco_data.TotalCharges, errors = 'coerce')
telco_data.isnull().sum()
```

3. As we can see there are 11 missing values in TotalCharges column. Let's check these records


```python
telco_data.loc[telco_data['TotalCharges'].isnull() == True]
```

Missing value treatement

Since the % of these records compared to total dataset is very low ie 0.15%, it is safe to ignore
them from further processing

```python
telco_data.dropna(how='any', inplace=True)
```

5. Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group
if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on

```python
print(telco_data['tenure'].max())
```

```python
# 1. Create clean labels for the tenure ranges
labels = ["{0} - {1}".format(i, i+11) for i in range(1, 72, 12)]

# 2. Create the tenure_group column using pd.cut
telco_data['tenure_group'] = pd.cut(
    telco_data['tenure'],          # column to bucket/bin
    bins=range(1, 80, 12),         # bin edges (1â€“12, 13â€“24, ..., 73â€“72)
    labels=labels,                # text labels for each bin
    right=False,                  # include the left edge, exclude the right
)

# 3. Display value counts for each group
print(telco_data['tenure_group'].value_counts())
```

6. Remove columns not required for processing

```python
telco_data.drop(columns= ['customerID','tenure'], axis=1,
inplace=True)
telco_data.head()
```

```python
telco_data['Churn'] = np.where(telco_data.Churn == 'Yes' , 1,0)
telco_data.head()
```

9. Relationship between Monthly Charges and Total Charges


```python
# Convert categorical columns into dummy/0-1 format
telco_data_dummies = pd.get_dummies(telco_data, drop_first=True)

# Convert the two columns to numeric (in case they are still strings)
telco_data_dummies['MonthlyCharges'] = pd.to_numeric(telco_data_dummies['MonthlyCharges'], errors='coerce')
telco_data_dummies['TotalCharges'] = pd.to_numeric(telco_data_dummies['TotalCharges'], errors='coerce')

# Drop any rows that have become NaN in these two columns
telco_data_dummies.dropna(subset=['MonthlyCharges', 'TotalCharges'], inplace=True)

```

Total Charges increase as Monthly Charges increase - as expected

10. Churn by Monthly Charges and Total Charges

```python
# Plot KDE for customers who did NOT churn
Mth = sns.kdeplot(
    telco_data_dummies.MonthlyCharges[telco_data_dummies['Churn'] == 0],
    color="red",
    shade=True
)

# Plot KDE for customers who churned
Mth = sns.kdeplot(
    telco_data_dummies.MonthlyCharges[telco_data_dummies['Churn'] == 1],
    ax=Mth,
    color="blue",
    shade=True
)

# Add legend and axis labels
Mth.legend(["No Churn", "Churn"], loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Monthly Charges')
Mth.set_title('Monthly charges by churn')

plt.show()
```

```python
Tot = sns.kdeplot(
    telco_data_dummies['TotalCharges'][telco_data_dummies['Churn'] == 0],
    color='red', shade=True
)


Tot = sns.kdeplot(
    telco_data_dummies['TotalCharges'][telco_data_dummies['Churn'] == 1],
    ax=Tot, color='blue', shade=True
)


Tot.legend(['No Churn','Churn'], loc='upper right')
Tot.set_ylabel('Density')
Tot.set_xlabel('Total Charges')
Tot.set_title('Total Charges by Churn')

plt.show()
```

Surprising insight as higher Churn at lower Total Charges
However if we combine the insights of 3 parameters i.e. Tenure, Monthly Charges & Total
Charges then the picture is bit clear :- Higher Monthly Charge at lower tenure results into lower
Total Charge. Hence, all these 3 factors viz Higher Monthly Charge, Lower tenure and Lower
Total Charge are linkd to High Churn.

```python
# Separate datasets
new_df1_target0 = telco_data[telco_data["Churn"] == 0]    # Non-Churned
new_df1_target1 = telco_data[telco_data["Churn"] == 1]    # Churned

# Improved plotting function
def uniplot(df, col, title, hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 16
    plt.rcParams['axes.titlesize'] = 20
    
    fig, ax = plt.subplots()
    width = len(df[col].unique()) + 6 + (3 * (len(df[hue].unique()) if hue else 1))
    fig.set_size_inches(width, 6)
    
    plt.xticks(rotation=30)
    plt.title(title)

    sns.countplot(
        data=df,
        x=col,
        order=df[col].value_counts().index,
        hue=hue,
        palette='bright',
        ax=ax
    )

    
    plt.show()


# Plot churned customers: Partner distribution split by gender
uniplot(
    df=new_df1_target1,
    col='Partner',
    title='Partner-wise Distribution for Churned Customers (split by Gender)',
    hue='gender'
)
```
